# Drop Unnamed column

```python
df= df.drop("Unnamed: 0",axis=1)
df.head()
```

# Change weekly study hours column

```python
df["WklyStudyHours"]= df["WklyStudyHours"].str.replace("05-Oct","5-10")
df.head()
```

# Gender Distribution

```python
mpl.figure(figsize=(5,6))
ax=sn.countplot(data=df,x="Gender")
ax.bar_label(ax.containers[0])
mpl.title("count of Gender")
mpl.show()
```

```python
gb=df.groupby("ParentEduc").agg({"MathScore":'mean',"ReadingScore":'mean',"WritingScore":'mean'})
gb
```

```python
mpl.figure(figsize=(5,6))
mpl.title("Relation between parents education and student score")
sn.heatmap(gb, annot = True)
mpl.show()
```

```python
# From the above chart we have concluded that the education of the parents have a good impact on thir scores.
```

```python
gb1=df.groupby("ParentMaritalStatus").agg({"MathScore":'mean',"ReadingScore":'mean',"WritingScore":'mean'})
gb1
```

```python
mpl.figure(figsize=(5,6))
sn.heatmap(gb1, annot = True)
mpl.title("Relation between parents marital Status and student score")
mpl.show()
```

```python
# by analyzing the above chart we observed that the impact of parents marital status have no/negligible impact on students score.
```

# Distribution of Ethnic Group

```python
groupA= df.loc[(df["EthnicGroup"]== "group A")].count()
groupB= df.loc[(df["EthnicGroup"]== "group B")].count()
groupC= df.loc[(df["EthnicGroup"]== "group C")].count()
groupD= df.loc[(df["EthnicGroup"]== "group D")].count()
groupE= df.loc[(df["EthnicGroup"]== "group E")].count()

l=["group A","group B","group C","group D","group E"]
mlist=[groupA["EthnicGroup"],groupB["EthnicGroup"],groupC["EthnicGroup"],groupD["EthnicGroup"],groupE["EthnicGroup"]]

mpl.pie(mlist,labels=l,autopct="%1.2f%%")
mpl.title("Distribution of Ethnic Groups")
mpl.show()
```

```python
ax=sn.countplot(data=df, x='EthnicGroup')
ax.bar_label(ax.containers[0])
```

```python
wh=df.groupby("WklyStudyHours").agg({"MathScore":'mean',"ReadingScore":'mean',"WritingScore":'mean'})
wh

mpl.figure(figsize=(5,6))
mpl.title("Relation between Weekly Study hour and student score")
sn.heatmap(wh, annot = True)
mpl.show()
```
